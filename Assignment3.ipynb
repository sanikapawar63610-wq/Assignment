# Section A - Classes and Objects

```python
"""
1. Create a class BankAccount with the following:

a) Attributes: account_holder, account_number, and balance.

b) Methods: deposit(amount), withdraw(amount), and get_balance().

c) withdraw() should print a message and do nothing if the amount exceeds the current balance.

d) A_str_method that returns a readable summary of the account.

Create two BankAccount objects and perform a few deposits and withdrawals on each.
""";
```

```python
class BankAccount:
    def __init__(self, account_holder, account_number, balance):
        self.account_holder = account_holder
        self.account_number = account_number
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(f"Deposited ₹{amount}")

    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient balance! Withdrawal not possible.")
        else:
            self.balance -= amount
            print(f"Withdrawn ₹{amount}")

    def get_balance(self):
        return self.balance

    def __str__(self):
        return (f"Account Holder: {self.account_holder}\n"
                f"Account Number: {self.account_number}\n"
                f"Balance: ₹{self.balance}")


# Create two BankAccount objects
acc1 = BankAccount("Rahul", 1001, 5000)
acc2 = BankAccount("Priya", 1002, 8000)

# Perform transactions
acc1.deposit(2000)
acc1.withdraw(3000)
acc1.withdraw(5000)

print("\n", acc1)

acc2.deposit(1000)
acc2.withdraw(2500)

print("\n", acc2)
```

```python
"""
2. Create a class Student with attributes name, roll_no, and marks (a list of 5 subject marks). Add methods:

a) calculate_percentage() - returns the average of all marks.

b) get_grade() - returns 'A' for >= 80, 'B' for >= 60, 'C' for >= 40, and 'Fail' otherwise.

c) display() prints all details including percentage and grade.

Create a list of 3 Student objects and call display() on each.
""";
```

```python
class Student:
    def __init__(self, name, roll_no, marks):
        self.name = name
        self.roll_no = roll_no
        self.marks = marks

    def calculate_percentage(self):
        return sum(self.marks) / len(self.marks)

    def get_grade(self):
        percentage = self.calculate_percentage()

        if percentage >= 80:
            return 'A'
        elif percentage >= 60:
            return 'B'
        elif percentage >= 40:
            return 'C'
        else:
            return 'Fail'

    def display(self):
        print("Name:", self.name)
        print("Roll No:", self.roll_no)
        print("Marks:", self.marks)
        print("Percentage:", self.calculate_percentage())
        print("Grade:", self.get_grade())
        print("-" * 30)


# Create list of 3 students
students = [
    Student("Amit", 1, [85, 90, 80, 88, 92]),
    Student("Sneha", 2, [65, 70, 60, 68, 72]),
    Student("Rohan", 3, [35, 40, 38, 42, 39])
]

# Display details
for student in students:
    student.display()
```

# Section B - Encapsulation

```python
"""
3. Create a class Employee with private attributes_name, _ emp_id, and_salary. Add:

a) Getter and setter methods for all three attributes.

b) The salary setter should raise a ValueError if the value is below 10000.

c) A method give_raise(percent) that increases salary by the given percentage.

Demonstrate that the private attributes cannot be accessed directly, and that all access goes through the methods.
""";
```

```python
class Employee:
    def __init__(self, name, emp_id, salary):
        self.__name = name
        self.__emp_id = emp_id
        self.__salary = salary

    # Getter methods
    def get_name(self):
        return self.__name

    def get_emp_id(self):
        return self.__emp_id

    def get_salary(self):
        return self.__salary

    # Setter methods
    def set_name(self, name):
        self.__name = name

    def set_emp_id(self, emp_id):
        self.__emp_id = emp_id

    def set_salary(self, salary):
        if salary < 10000:
            raise ValueError("Salary cannot be less than 10000")
        self.__salary = salary

    # Method to increase salary
    def give_raise(self, percent):
        self.__salary += self.__salary * percent / 100


# Create Employee object
emp = Employee("Rahul", 101, 25000)

# Access using getter methods
print("Name:", emp.get_name())
print("Employee ID:", emp.get_emp_id())
print("Salary:", emp.get_salary())

# Give 10% raise
emp.give_raise(10)
print("Salary after raise:", emp.get_salary())

# Set new salary
emp.set_salary(30000)
print("Updated Salary:", emp.get_salary())

# Demonstrate ValueError
try:
    emp.set_salary(8000)
except ValueError as e:
    print(e)

# Direct access to private attribute
try:
    print(emp.__salary)
except AttributeError:
    print("Private attributes cannot be accessed directly.")
```

```python
"""
4. Build a class SecureVault that stores a private_pin and a private_data dictionary. It should have:

a) A method add_entry(pin, key, value) that adds to_data only if the pin matches.

b) A method get_entry(pin, key) that returns the value only if the pin matches.

c) Any wrong pin attempt should print 'Access Denied' and return nothing.
""";
```

```python
class SecureVault:
    def __init__(self, pin):
        self.__pin = pin
        self.__data = {}

    def add_entry(self, pin, key, value):
        if pin == self.__pin:
            self.__data[key] = value
            print("Entry added successfully.")
        else:
            print("Access Denied")

    def get_entry(self, pin, key):
        if pin == self.__pin:
            return self.__data.get(key, "Key not found")
        else:
            print("Access Denied")
            return None


# Create SecureVault object
vault = SecureVault("1234")

# Add entries
vault.add_entry("1234", "Username", "Dhanshri")
vault.add_entry("1234", "Password", "abc123")

# Retrieve entry with correct PIN
print("Username:", vault.get_entry("1234", "Username"))

# Wrong PIN attempts
vault.add_entry("1111", "Email", "user@gmail.com")
print(vault.get_entry("1111", "Password"))
```

# Section C - Inheritance and Method Overriding

```python
"""
5. Create a base class Shape with a method area() that returns 0 and a method describe() 
that prints 'I am a shape'. Create three subclasses:

a) Circle(radius) - overrides area() using pi*r^2.

b) Rectangle(length, width) - overrides area().

c) Triangle(base, height) overrides area().

Each subclass should also override describe() to print something specific about itself. 
Create one object of each and call both methods.
""";
```

```python
import math

# Base class
class Shape:
    def area(self):
        return 0

    def describe(self):
        print("I am a shape")


# Subclass: Circle
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2

    def describe(self):
        print("I am a Circle")


# Subclass: Rectangle
class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def describe(self):
        print("I am a Rectangle")


# Subclass: Triangle
class Triangle(Shape):
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area(self):
        return 0.5 * self.base * self.height

    def describe(self):
        print("I am a Triangle")


# Create objects
circle = Circle(7)
rectangle = Rectangle(10, 5)
triangle = Triangle(8, 6)

# Call methods
circle.describe()
print("Area =", circle.area())

rectangle.describe()
print("Area =", rectangle.area())

triangle.describe()
print("Area =", triangle.area())
```

```python
"""
6. Create a class Vehicle with attributes make, model, and year and a method start_engine() 
that prints 'Engine started". Create two subclasses:

a) Car(make, model, year, num_doors) - overrides start_engine() to include the car's name in the message.

b) ElectricCar(make, model, year, battery_capacity) - overrides start_engine() 
to print 'Silent start- battery at X kWh'.

Use super() in both subclasses to call the parent_init. Create objects and test all methods.
""";
```

```python
# Base class
class Vehicle:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    def start_engine(self):
        print("Engine started")


# Subclass: Car
class Car(Vehicle):
    def __init__(self, make, model, year, num_doors):
        super().__init__(make, model, year)   # Call parent constructor
        self.num_doors = num_doors

    def start_engine(self):
        print(f"Engine started for {self.make} {self.model}")


# Subclass: ElectricCar
class ElectricCar(Vehicle):
    def __init__(self, make, model, year, battery_capacity):
        super().__init__(make, model, year)   # Call parent constructor
        self.battery_capacity = battery_capacity

    def start_engine(self):
        print(f"Silent start - battery at {self.battery_capacity} kWh")


# Create objects
car = Car("Toyota", "Fortuner", 2023, 4)
electric_car = ElectricCar("Tesla", "Model 3", 2024, 75)

# Display details
print("Car Details:")
print(car.make, car.model, car.year, car.num_doors)
car.start_engine()

print("\nElectric Car Details:")
print(electric_car.make, electric_car.model, electric_car.year, electric_car.battery_capacity)
electric_car.start_engine()
```

# Section D - Polymorphism

```python
"""
7. Create a base class Animal with a method sound() that prints 'Some sound'. 
Create subclasses Dog, Cat, and Cow, each overriding sound() with their respective sounds.
Write a function make_all_sounds(animals) that takes a list of Animal objects and calls sound() on each. 
Demonstrate that the correct method is called regardless of the object type.
""";
```

```python
# Base class
class Animal:
    def sound(self):
        print("Some sound")


# Subclass: Dog
class Dog(Animal):
    def sound(self):
        print("Dog says: Woof! Woof!")


# Subclass: Cat
class Cat(Animal):
    def sound(self):
        print("Cat says: Meow! Meow!")


# Subclass: Cow
class Cow(Animal):
    def sound(self):
        print("Cow says: Moo! Moo!")


# Function to call sound() on each animal
def make_all_sounds(animals):
    for animal in animals:
        animal.sound()


# Create objects
dog = Dog()
cat = Cat()
cow = Cow()

# Store objects in a list
animals = [dog, cat, cow]

# Call the function
make_all_sounds(animals)
```

```python
"""
8. Create a class Calculator with overloaded-style methods using default arguments:



a) add(a, b, c=0) - adds two or three numbers.

b) multiply(a, b, c=1) - multiplies two or three numbers.

c) Also implement_add_ on a class Vector(x, y) 
so that two Vector objects can be added with the + operator. Print the resulting vector.
""";
```

```python
# Calculator class
class Calculator:

    # Add two or three numbers
    def add(self, a, b, c=0):
        return a + b + c

    # Multiply two or three numbers
    def multiply(self, a, b, c=1):
        return a * b * c


# Vector class
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    # Overload + operator
    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    # String representation
    def __str__(self):
        return f"({self.x}, {self.y})"


# Create Calculator object
calc = Calculator()

print("Addition of two numbers:", calc.add(10, 20))
print("Addition of three numbers:", calc.add(10, 20, 30))

print("Multiplication of two numbers:", calc.multiply(5, 4))
print("Multiplication of three numbers:", calc.multiply(5, 4, 2))

# Create Vector objects
v1 = Vector(2, 3)
v2 = Vector(4, 5)

# Add vectors
v3 = v1 + v2

print("Vector 1:", v1)
print("Vector 2:", v2)
print("Resultant Vector:", v3)
```

# Section E - File Handling

```python
"""
9. Perform the following file operations. Use plain text files only.

a) Write a list of 5 student names to a file students.txt, one name per line.

b) Read the file back and print each name with its line number.

c) Append two more names to the same file without overwriting existing content.

d) Read the final file and count the total number of students.

e) Copy the contents of students.txt into a new file students_backup.txt.
""";
```

```python
students = ["Rahul", "Priya", "Amit", "Sneha", "Rohan"]

with open("students.txt", "w") as file:
    for name in students:
        file.write(name + "\n")

print("5 student names written to students.txt")

# b) Read the file and print each name with line number
print("\nStudents with Line Numbers:")
with open("students.txt", "r") as file:
    for line_no, name in enumerate(file, start=1):
        print(f"{line_no}. {name.strip()}")

# c) Append two more names
with open("students.txt", "a") as file:
    file.write("Neha\n")
    file.write("Karan\n")

print("\nTwo more names appended.")

# d) Read the final file and count total students
with open("students.txt", "r") as file:
    names = file.readlines()

print("\nFinal Student List:")
for name in names:
    print(name.strip())

print("Total number of students:", len(names))

# e) Copy contents to students_backup.txt
with open("students.txt", "r") as source:
    content = source.read()

with open("students_backup.txt", "w") as backup:
    backup.write(content)

print("\nContents copied to students_backup.txt")
```

# Section F - Menu-Driven App with OOP

```python
"""
10. Build a menu-driven Student Record Manager using OOP and file handling.
The app should have a class StudentManager with the following menu options:

1. Add Student - takes name, roll number, and marks for 3 subjects.

2. View All - displays all stored students with their average marks.

3. Search Student - finds a student by roll number.

4. Delete Student removes a student by roll number.

5. Save to File - writes all records to students_data.txt.

6. Load from File - reads records back from students_data.txt on startup.

7. Exit

Requirements: use a class with a list as internal storage, handle invalid menu choices gracefully,
and use a while loop to keep the menu running until the user chooses to exit.
""";
```

```python
class StudentManager:
    def __init__(self):
        self.students = []

    # Load data from file
    def load_from_file(self):
        try:
            with open("students_data.txt", "r") as file:
                for line in file:
                    data = line.strip().split(",")
                    if len(data) == 5:
                        student = {
                            "name": data[0],
                            "roll": data[1],
                            "marks": [int(data[2]), int(data[3]), int(data[4])]
                        }
                        self.students.append(student)
            print("Student records loaded successfully.")
        except FileNotFoundError:
            print("No existing file found. Starting with empty records.")

    # Save data to file
    def save_to_file(self):
        with open("students_data.txt", "w") as file:
            for student in self.students:
                file.write(f"{student['name']},{student['roll']},{student['marks'][0]},{student['marks'][1]},{student['marks'][2]}\n")
        print("Records saved successfully.")

    # Add student
    def add_student(self):
        name = input("Enter Name: ")
        roll = input("Enter Roll Number: ")

        marks = []
        for i in range(3):
            marks.append(int(input(f"Enter Marks for Subject {i+1}: ")))

        self.students.append({
            "name": name,
            "roll": roll,
            "marks": marks
        })
        print("Student added successfully.")

    # View all students
    def view_all(self):
        if not self.students:
            print("No student records found.")
            return

        print("\nStudent Records")
        for student in self.students:
            avg = sum(student["marks"]) / 3
            print("-----------------------------")
            print("Name :", student["name"])
            print("Roll :", student["roll"])
            print("Marks:", student["marks"])
            print("Average:", avg)

    # Search student
    def search_student(self):
        roll = input("Enter Roll Number to Search: ")
        for student in self.students:
            if student["roll"] == roll:
                avg = sum(student["marks"]) / 3
                print("Student Found")
                print("Name :", student["name"])
                print("Marks:", student["marks"])
                print("Average:", avg)
                return
        print("Student not found.")

    # Delete student
    def delete_student(self):
        roll = input("Enter Roll Number to Delete: ")
        for student in self.students:
            if student["roll"] == roll:
                self.students.remove(student)
                print("Student deleted successfully.")
                return
        print("Student not found.")


# Main Program
manager = StudentManager()
manager.load_from_file()

while True:
    print("\n===== Student Record Manager =====")
    print("1. Add Student")
    print("2. View All")
    print("3. Search Student")
    print("4. Delete Student")
    print("5. Save to File")
    print("6. Load from File")
    print("7. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        manager.add_student()

    elif choice == "2":
        manager.view_all()

    elif choice == "3":
        manager.search_student()

    elif choice == "4":
        manager.delete_student()

    elif choice == "5":
        manager.save_to_file()

    elif choice == "6":
        manager.load_from_file()

    elif choice == "7":
        manager.save_to_file()
        print("Thank You! Exiting Program.")
        break

    else:
        print("Invalid choice! Please enter a number between 1 and 7.")
```
