```python
"""
1. Write a function analyze_sentence(sentence) that takes a sentence and returns
a dictionary with word_count, char_count (excluding spaces), 
unique_words (case-insensitive), and most_frequent_word. Test it with:

kpoints

"The quick brown fox jumps over the lazy dog the fox"
""";
```

```python
def analyze_sentence(sentence):
    words = sentence.lower().split()

    result = {
        "word_count": len(words),
        "char_count": len(sentence.replace(" ", "")),
        "unique_words": len(set(words))
    }

    freq = {}
    for word in words:
        freq[word] = freq.get(word, 0) + 1

    most = max(freq, key=freq.get)
    result["most_frequent_word"] = most

    return result


sentence = "The quick brown fox jumps over the lazy dog the fox"

print(analyze_sentence(sentence))
```

```python
"""
2. Given the string below, do the following and print each result:

text "A man a plan a canal Panama"

a) Reverse the string without using slicing.

b) Check if it is a palindrome, ignoring case and spaces.

weekly Assignment.pdf

md

c) Replace every vowel with ".

d) Count how many times each vowel appears.
""";
```

```python
text = "A man a plan a canal Panama"

# (a) Reverse without slicing
reverse = ""
for ch in text:
    reverse = ch + reverse
print("Reversed:", reverse)

# (b) Palindrome
s = text.replace(" ", "").lower()
if s == "".join(reversed(s)):
    print("Palindrome")
else:
    print("Not Palindrome")

# (c) Replace vowels with *
vowels = "aeiouAEIOU"
new_text = ""
for ch in text:
    if ch in vowels:
        new_text += "*"
    else:
        new_text += ch
print("After replacing vowels:", new_text)

# (d) Count vowels
count = {'a':0,'e':0,'i':0,'o':0,'u':0}

for ch in text.lower():
    if ch in count:
        count[ch] += 1

print("Vowel Count:", count)
```

```python
"""
3. Write a function format_report(data) that takes a lift of tuples (name, score, grade) and prints a
formatted table where name is left-aligned (width 15), score is right-aligned (width 6),
and grade is centered (width 8). Use this data: data [("Alice", 92, "A"), ("Bob", 78, "B+"),
("Charlie", 55, "C"), ("Diana", 100, "A+")]
""";
```

```python
def format_report(data):
    print(f'{"Name":<15}{"Score":>6}{"Grade":^8}')
    print("-" * 29)

    for name, score, grade in data:
        print(f'{name:<15}{score:>6}{grade:^8}')


data = [
    ("Alice", 92, "A"),
    ("Bob", 78, "B+"),
    ("Charlie", 55, "C"),
    ("Diana", 100, "A+")
]

format_report(data)
```

```python
"""
4. Write a function list_ops(nums) that:

a) Removes duplicates while preserving the original order.

b) Returns a new list with only elements at even indices, squared.

c) Flattens the nested list into a single list using list comprehension.

nums [3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5] nested [[1, 2, 3], [4, 5], [6, 7, 8,91]
""";
```

```python
def list_ops(nums):
    # (a) Remove duplicates
    unique = []
    for i in nums:
        if i not in unique:
            unique.append(i)

    # (b) Square elements at even indices
    squared = [nums[i]**2 for i in range(0, len(nums), 2)]

    # (c) Flatten nested list
    nested = [[1, 2, 3], [4, 5], [6, 7, 8, 9]]
    flat = [item for sublist in nested for item in sublist]

    print("Unique List:", unique)
    print("Squared Even Index Elements:", squared)
    print("Flattened List:", flat)


nums = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5]
list_ops(nums)
```

```python
"""
7. Write a function inventory_manager(records) that takes a list of tuples
(product, category, quantity, price) and returns a dictionary grouped by category.
Each category should have total_items, total_value, and a list of product names. Use:

records = [ ("Laptop", "Electronics", 10, 55000), 
("Mouse", "Electronics", 50, 499), ("Desk", "Furniture", 8, 12000), 
("Chair", "Furniture", 15, 4500), ("Pen", "Stationery", 200, 20), 1
""";
```

```python
def inventory_manager(records):
    inventory = {}

    for product, category, quantity, price in records:
        if category not in inventory:
            inventory[category] = {
                "total_items": 0,
                "total_value": 0,
                "products": []
            }

        inventory[category]["total_items"] += quantity
        inventory[category]["total_value"] += quantity * price
        inventory[category]["products"].append(product)

    return inventory


records = [
    ("Laptop", "Electronics", 10, 55000),
    ("Mouse", "Electronics", 50, 499),
    ("Desk", "Furniture", 8, 12000),
    ("Chair", "Furniture", 15, 4500),
    ("Pen", "Stationery", 200, 20)
]

result = inventory_manager(records)

for category, details in result.items():
    print(category, ":", details)
```

```python
"""
8. Implement the following:

a) A function power_all(*args, exp=2) that returns a list of each number raised to exp.

b) A function profile(**kwargs) that prints each key-value pair as Key: Value.

c) Use a lambda inside sorted() to sort the list of dicts below by cgpa in descending order.

students [{"name": "Arjun", "cgpa": 8.4), {"name": "Sanya", "cgpa": 9.1}, {"name": "Dev", "cgpa": 7.8}]
""";
```

```python
def power_all(*args, exp=2):
    return [num ** exp for num in args]

print(power_all(2, 3, 4))
print(power_all(2, 3, 4, exp=3))
```

```python
def profile(**kwargs):
    for key, value in kwargs.items():
        print(key, ":", value)

profile(name="Arjun", age=21, city="Pune")
```

```python
"""
9. Create a module stats_utils.py with three functions:

a) mean(data) - arithmetic mean.

b) variance(data) - population variance without using the statistics module.

c) normalize(data)-min-max normalization returning values between 0 and 1.

In a separate main.py, import and use these functions on the dataset below. 
Also compute the standard deviation using the math module. dataset [12, 45, 7, 23, 56, 89, 34, 67, 11, 90]
""";
```

```python
def mean(data):
    return sum(data) / len(data)


def variance(data):
    m = mean(data)
    return sum((x - m) ** 2 for x in data) / len(data)


def normalize(data):
    minimum = min(data)
    maximum = max(data)
    return [(x - minimum) / (maximum - minimum) for x in data]
```

```python
import math
import stats_utils

dataset = [12, 45, 7, 23, 56, 89, 34, 67, 11, 90]

print("Mean =", stats_utils.mean(dataset))

var = stats_utils.variance(dataset)
print("Variance =", var)

print("Normalized Data =", stats_utils.normalize(dataset))

std = math.sqrt(var)
print("Standard Deviation =", std)
```

```python
"""
10. Complete the following:

a) Write a closure make_multiplier(n) that returns a function multiplying its input by n. Create triple and quintuple using it.

b) Write a decorator @timer that prints the execution time of any function it wraps. Apply it to slow_sum(n) which computes the sum of squares from 1 to n.

c) Write a decorator @validate_positive that raises a ValueError if any argument is zero or negative. Apply it to safe_sqrt(x).
""";
```

```python
def make_multiplier(n):
    def multiply(x):
        return x * n
    return multiply

triple = make_multiplier(3)
quintuple = make_multiplier(5)

print(triple(10))
print(quintuple(10))
```

```python
import time

def timer(func):
    def wrapper(n):
        start = time.time()
        result = func(n)
        end = time.time()
        print("Execution Time:", end - start, "seconds")
        return result
    return wrapper


@timer
def slow_sum(n):
    total = 0
    for i in range(1, n + 1):
        total += i * i
    return total

print("Sum =", slow_sum(100000))
```

```python
import math

def validate_positive(func):
    def wrapper(x):
        if x <= 0:
            raise ValueError("Input must be positive")
        return func(x)
    return wrapper


@validate_positive
def safe_sqrt(x):
    return math.sqrt(x)

print(safe_sqrt(25))

# Uncomment to test error
# print(safe_sqrt(-9))
```
