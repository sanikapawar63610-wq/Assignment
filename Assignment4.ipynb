# Section A - Exception Handling

```python
"""
1. Write a function safe_divide(a, b) that divides two numbers and handles 
ZeroDivisionError and TypeError separately, printing an appropriate message for each. 
Use a finally block to print 'Operation attempted.' regardless of the outcome.
Test it with valid numbers, division by zero, and a string input.
""";
```

```python
def safe_divide(a, b):
    try:
        result = a / b
        print("Result:", result)

    except ZeroDivisionError:
        print("Error: Cannot divide by zero.")

    except TypeError:
        print("Error: Both inputs must be numbers.")

    finally:
        print("Operation attempted.")

# Test Cases

print("Test 1: Valid Division")
safe_divide(20, 5)

print("\nTest 2: Division by Zero")
safe_divide(10, 0)

print("\nTest 3: String Input")
safe_divide("20", 5)
```

```python
"""
2. Write a function validate_age(age) that raises a custom exception InvalidAgeError 
(a class you define, inheriting from Exception) if age is negative or greater than 120. 
Catch this exception in a calling block and print a user-friendly message.
Test it with at least 3 different values.
""";
```

```python
# Custom Exception
class InvalidAgeError(Exception):
    pass

# Function to validate age
def validate_age(age):
    if age < 0 or age > 120:
        raise InvalidAgeError("Age must be between 0 and 120.")
    else:
        print("Valid age:", age)

# Test Cases
ages = [25, -5, 150]

for age in ages:
    try:
        validate_age(age)
    except InvalidAgeError as e:
        print("Invalid Age Error:", e)
```

```python
"""
3. Write a program that repeatedly asks the user (use a loop with at least 3 simulated inputs in a list, e.g.
user_inputs = ['10', 'abc', '0', '5']) to enter a number and divide 100 by it. 
Handle the following in a single try block using multiple except clauses:

a) ValueError if the input is not a number.

b) ZeroDivisionError if the input is zero.

c) A generic Exception as a catch-all for anything else.

Print the result if successful, otherwise print which error occurred.
""";
```

```python
# Simulated user inputs
user_inputs = ['10', 'abc', '0', '5']

for user_input in user_inputs:
    print("Input:", user_input)

    try:
        num = int(user_input)
        result = 100 / num
        print("Result:", result)

    except ValueError:
        print("Error: Invalid number entered.")

    except ZeroDivisionError:
        print("Error: Cannot divide by zero.")

    except Exception as e:
        print("Unexpected Error:", e)

    print()
```

# Section B - Numpy

```python
"""
4. Create a 1D NumPy array of integers from 1 to 20. Perform the following:

a) Reshape it into a 4x5 2D array.

b) Print the sum of each row and each column.

c) Extract all elements greater than 10 using boolean indexing.

d) Replace all even numbers in the array with -1.
""";
```

```python
import numpy as np

# Create a 1D array from 1 to 20
arr = np.arange(1, 21)

print("Original 1D Array:")
print(arr)

# a) Reshape into 4x5 array
arr2 = arr.reshape(4, 5)
print("\n4x5 Array:")
print(arr2)

# b) Sum of each row and each column
print("\nSum of each row:")
print(arr2.sum(axis=1))

print("\nSum of each column:")
print(arr2.sum(axis=0))

# c) Elements greater than 10
print("\nElements greater than 10:")
print(arr2[arr2 > 10])

# d) Replace all even numbers with -1
arr2[arr2 % 2 == 0] = -1

print("\nArray after replacing even numbers with -1:")
print(arr2)
```

```python
"""
5. Create two NumPy arrays A and B as shown below and perform:

Anp.array([[1, 2], [3, 4]]) B= np.array([[5, 6], [7, 8]])

a) Element-wise addition, subtraction, and multiplication.

b) Matrix multiplication using np.dot() or the @ operator.

c) Transpose of A.

d) The determinant and inverse of A using np.linalg.
""";
```

```python
import numpy as np

# Create arrays
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

print("Matrix A:")
print(A)

print("\nMatrix B:")
print(B)

# a) Element-wise operations
print("\nElement-wise Addition:")
print(A + B)

print("\nElement-wise Subtraction:")
print(A - B)

print("\nElement-wise Multiplication:")
print(A * B)

# b) Matrix Multiplication
print("\nMatrix Multiplication (A @ B):")
print(np.dot(A, B))
# or print(A @ B)

# c) Transpose of A
print("\nTranspose of A:")
print(A.T)

# d) Determinant and Inverse of A
print("\nDeterminant of A:")
print(np.linalg.det(A))

print("\nInverse of A:")
print(np.linalg.inv(A))
```

```python
"""
6. Generate a NumPy array of 15 random integers between 1 and 100 using np.random.randint() 
(set a seed of 42 for reproducibility). Then:

a) Find the mean, median, standard deviation, min, and max.

b) Sort the array in descending order.

c) Find the indices of the 3 largest values using np.argsort().

d) Normalize the array so all values lie between 0 and 1.
""";
```

```python
import numpy as np

# Set seed for reproducibility
np.random.seed(42)

# Generate 15 random integers between 1 and 100
arr = np.random.randint(1, 101, 15)

print("Original Array:")
print(arr)

# a) Mean, Median, Standard Deviation, Min, Max
print("\nMean:", np.mean(arr))
print("Median:", np.median(arr))
print("Standard Deviation:", np.std(arr))
print("Minimum:", np.min(arr))
print("Maximum:", np.max(arr))

# b) Sort the array in descending order
sorted_arr = np.sort(arr)[::-1]
print("\nArray in Descending Order:")
print(sorted_arr)

# c) Indices of the 3 largest values
indices = np.argsort(arr)[-3:][::-1]
print("\nIndices of the 3 Largest Values:")
print(indices)

# d) Normalize the array between 0 and 1
normalized = (arr - np.min(arr)) / (np.max(arr) - np.min(arr))
print("\nNormalized Array:")
print(normalized)
```

# Section C - Pandas Basics

```python
"""
7. Create a Pandas DataFrame from the dictionary below and perform the following:

data {"Name": ["Aarav", "Priya", "Karan", "Meera", "Sahil"], "Age": [23, 21, 25, 22, 24], "Department":
["AI", "Data Science", "AI", "Web Dev", "Data Science"], "Score": [88, 92, 79, 95, 67], }

a) Display the first 3 rows and the data types of each column.

b) Add a new column Grade based on Score (>=90 'A', >=75 'B', else 'C').

c) Sort the DataFrame by Score in descending order.

d) Filter rows where Department is 'Al'.
""";
```

```python
import pandas as pd

# Create DataFrame
data = {
    "Name": ["Aarav", "Priya", "Karan", "Meera", "Sahil"],
    "Age": [23, 21, 25, 22, 24],
    "Department": ["AI", "Data Science", "AI", "Web Dev", "Data Science"],
    "Score": [88, 92, 79, 95, 67]
}

df = pd.DataFrame(data)

# a) Display first 3 rows and data types
print("First 3 Rows:")
print(df.head(3))

print("\nData Types:")
print(df.dtypes)

# b) Add Grade column
def grade(score):
    if score >= 90:
        return "A"
    elif score >= 75:
        return "B"
    else:
        return "C"

df["Grade"] = df["Score"].apply(grade)

print("\nDataFrame with Grade:")
print(df)

# c) Sort by Score in descending order
sorted_df = df.sort_values(by="Score", ascending=False)

print("\nSorted by Score (Descending):")
print(sorted_df)

# d) Filter rows where Department is 'AI'
ai_df = df[df["Department"] == "AI"]

print("\nStudents in AI Department:")
print(ai_df)
```

```python
"""
8. Using the same DataFrame from Q7:

a) Group by Department and find the average Score for each.

b) Find the row with the highest Score using idxmax().

c) Rename the column Score to Marks.

d) Drop the Age column and display the resulting DataFrame.
""";
```

```python
import pandas as pd

# Create DataFrame
data = {
    "Name": ["Aarav", "Priya", "Karan", "Meera", "Sahil"],
    "Age": [23, 21, 25, 22, 24],
    "Department": ["AI", "Data Science", "AI", "Web Dev", "Data Science"],
    "Score": [88, 92, 79, 95, 67]
}

df = pd.DataFrame(data)

# a) Group by Department and find average Score
print("Average Score by Department:")
print(df.groupby("Department")["Score"].mean())

# b) Find the row with the highest Score using idxmax()
print("\nStudent with Highest Score:")
print(df.loc[df["Score"].idxmax()])

# c) Rename 'Score' column to 'Marks'
df = df.rename(columns={"Score": "Marks"})
print("\nDataFrame after Renaming:")
print(df)

# d) Drop the Age column
df = df.drop(columns=["Age"])
print("\nDataFrame after Dropping Age:")
print(df)
```

```python
"""
9. Create a DataFrame with some missing values as shown below and handle them:

import numpy as np data = { "Product": ["Pen", "Notebook", "Eraser", "Sharpener", "Ruler"],
"Price": [10, np.nan, 5, 8, np.nan], "Stock": [100, 50, np.nan, 30, 20], )

a) Check the number of missing values in each column using isnull().sum().

b) Fill missing Price values with the column mean.

c) Drop rows where Stock is still missing.

d) Save the cleaned DataFrame to a CSV file named cleaned_products.csv without the index.
""";
```

```python
import pandas as pd
import numpy as np

# Create DataFrame
data = {
    "Product": ["Pen", "Notebook", "Eraser", "Sharpener", "Ruler"],
    "Price": [10, np.nan, 5, 8, np.nan],
    "Stock": [100, 50, np.nan, 30, 20]
}

df = pd.DataFrame(data)

# a) Check missing values
print("Missing Values in Each Column:")
print(df.isnull().sum())

# b) Fill missing Price values with the column mean
df["Price"] = df["Price"].fillna(df["Price"].mean())

print("\nDataFrame after filling missing Price:")
print(df)

# c) Drop rows where Stock is missing
df = df.dropna(subset=["Stock"])

print("\nDataFrame after dropping missing Stock:")
print(df)

# d) Save the cleaned DataFrame to CSV
df.to_csv("cleaned_products.csv", index=False)

print("\nData saved successfully to 'cleaned_products.csv'")
```
