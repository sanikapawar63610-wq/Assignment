```python
"""
1. Take two numbers as input from the user and print their sum,
difference, product, and quotient each on a separate line. 
Handle division by zero with an appropriate message.
""";
```

```python
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

print("Sum =", num1 + num2)
print("Difference =", num1 - num2)
print("Product =", num1 * num2)

if num2 != 0:
    print("Quotient =", num1 / num2)
else:
    print("Division by zero is not possible.")
```

```python
"""
2. Accept a temperature in Celsius from the user, 
convert it to Fahrenheit and Kelvin, 
and print all three values with labels.
""";
```

```python
celsius = float(input("Enter temperature in Celsius: "))

fahrenheit = (celsius * 9/5) + 32
kelvin = celsius + 273.15

print("Celsius =", celsius)
print("Fahrenheit =", fahrenheit)
print("Kelvin =", kelvin)
```

```python
text = input("Enter a string: ")

print("Uppercase:", text.upper())
print("Length:", len(text))
print("Reversed:", text[::-1])
```

```python
"""
4. Take three numbers as input and print the largest among them. Do not use the built-in max() function.
""";
```

```python
a = float(input("Enter first number: "))
b = float(input("Enter second number: "))
c = float(input("Enter third number: "))

if a >= b and a >= c:
    largest = a
elif b >= a and b >= c:
    largest = b
else:
    largest = c

print("Largest number =", largest)
```

```python
"""
5. Take a year as input and print whether it is a leap year or not.
""";
```

```python
year = int(input("Enter a year: "))

if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    print(year, "is a Leap Year")
else:
    print(year, "is Not a Leap Year")
```

```python
"""
6. Build a simple calculator - take two numbers and an operator (+, -, *, /) as input,
perform the operation, and print the result.
Print an error message for invalid operators or division by zero.
""";
```

```python
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
op = input("Enter operator (+, -, *, /): ")

if op == "+":
    print("Result =", num1 + num2)
elif op == "-":
    print("Result =", num1 - num2)
elif op == "*":
    print("Result =", num1 * num2)
elif op == "/":
    if num2 != 0:
        print("Result =", num1 / num2)
    else:
        print("Error: Division by zero.")
else:
    print("Error: Invalid operator.")
```

```python
"""
8. Take a positive integer n as input and print all prime numbers from 2 to n.
""";
```

```python
n = int(input("Enter a positive integer: "))

for num in range(2, n + 1):
    prime = True
    for i in range(2, num):
        if num % i == 0:
            prime = False
            break
    if prime:
        print(num)
```

```python
"""
9. Keep accepting numbers from the user until they enter 0.
Then print the total count of numbers entered, their sum, 
and their average. Do not include 0 in the calculations.
""";
```

```python
count = 0
total = 0

while True:
    num = int(input("Enter a number (0 to stop): "))
    if num == 0:
        break
    count += 1
    total += num

if count > 0:
    print("Count =", count)
    print("Sum =", total)
    print("Average =", total / count)
else:
    print("No numbers entered.")
```

```python
"""
10. Take n as input and print a triangle pattern of n rows made of asterisks, 
where row i has (2i - 1) stars centered appropriately.
""";
```

```python
n = int(input("Enter number of rows: "))

for i in range(1, n + 1):
    print(" " * (n - i) + "*" * (2 * i - 1))
```
